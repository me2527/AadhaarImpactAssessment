### Context on Aadhaar

Aadhaar (Aadhaar card) is a 12-digit unique identity number (UIN) obtained by the residents of India, based on their biometric and demographic data.  

<center><img src="https://curriculum.whitehatjr.com/APT+Asset/APT+C4+images/adhar.jpg" width = 600></center>


**Objective of Aadhaar**

Aadhaar was a revolutionary step taken by the Indian government to identify residents and make beneficiary programs for them. World Bank Chief Economist Paul Romer described Aadhaar as "the most sophisticated ID program in the world."

On $23^{\text{rd}}$ September 2013, the Supreme court issued an interim order, **"No person should suffer for not getting Aadhaar"**, ordered that the government cannot deny a service to a resident for not having Aadhaar, preserving its voluntary nature. 

The Unique Identification Authority of India (UIDAI) is a statutory authority and a government department, under the Ministry of Electronics and Information Technology, following the provisions of the **Aadhaar Act 2016**. UIDAI is mandated to assign every citizen a $12$ digit Unique Identification Digit (UID) to all residents of the Indian republic. 

**Disputes on Aadhaar**

After the launch of the scheme, a lot of institutions were making Aadhaar card mandatory to avail of major necessary services such as taking admission in schools, opening a savings account in a bank, free ration for people below the poverty line.

In the September 2018 judgment, the Supreme Court gave an order that the Aadhaar card is **not mandatory** for opening bank accounts, getting a mobile number, or admission to the school. There has been resistance to Aadhaar from civil liberty groups as well due to privacy issues.

It should be noted that Aadhaar is just proof of identity and not of citizenship. Starting its services from $2010$ UIDAI has tried to issue UID to all residents of India. 


**Significance of Impact Assessment of Aadhaar**

After the Supreme court order, we want to find out whether the Aadhaar card is being used by the major institutions in the way it is intended to and people are not unnecessarily inconvenienced for not having it.

It has become an important subject of an investigation to understand and analyze which factors affect the issuing of UID and which do not. The impact of Aadhaar on the life of people, both positive and negative is also very intriguing. Hence, this becomes a very important problem statement of Data Science to deal with.

#### Activity 6: Sentiment Analysis

Perception of people about Aadhaar is one of the important parameter which is required to calculate its impact.

Let's observe whether people think Aadhaar have affected their life.



````
# Calculate sentiment effects of people concerning Aadhaar 
sentiment_effects_data=indepth_individual['xh_sentiment_affect_on_hh']
discard_null_values= sentiment_effects_data.dropna()
print("Sentiment Effects of People Concerning Aadhaar: ", discard_null_values.head(10))

count_effects=indepth_individual[(indepth_individual['xh_sentiment_affect_on_hh']=='Affects us a lot') ]
print("Affects us a lot", count_effects.head())

count_effects=indepth_individual[(indepth_individual['xh_sentiment_affect_on_hh']=='Affects us a little') ]
print("Affects us a little", count_effects.head())

count_effects=indepth_individual[(indepth_individual['xh_sentiment_affect_on_hh']=="Don't Know") ]
print("Don't Know", count_effects.head())

count_effects=indepth_individual[(indepth_individual['xh_sentiment_affect_on_hh']=="Doesn't affect us at all") ]
print(" Doesn't affect us at all", count_effects.head())
````



Observe whether people are satisfied with the scheme.


````
# Calculate sentiment satisfication of Aadhaar 
sentiment_sat_data=indepth_individual['xh_sentiment_satisfaction_aadhaar']
discard_null_values= sentiment_sat_data.dropna()
print("Sentiment Satisfication of Aadhaar: ", discard_null_values)

count_satisfied=indepth_individual[(indepth_individual['xh_sentiment_satisfaction_aadhaar']=="Satisfied") ]
print(" Satisfied", count_satisfied.head())


count_dont_know=indepth_individual[(indepth_individual['xh_sentiment_satisfaction_aadhaar']=="Don't Know") ]
print(" Don't Know", count_dont_know.head())

count_neither_sat_unsat=indepth_individual[(indepth_individual['xh_sentiment_satisfaction_aadhaar']=="Neither satisfied nor unsatisfied") ]
print(" Neither satisfied nor unsatisfied", count_neither_sat_unsat.head())
````
